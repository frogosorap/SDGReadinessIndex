# Sustainable Development Goal 3: Good Health and Well-Being

This Jupyter Notebook focuses on the development and preparation of the SDG 3 Sub-Index, which relates to the United Nations Sustainable Development Goal 3: Good Health and Well-Being.

The purpose of this notebook is to import, inspect, clean, and prepare datasets associated with health and well-being indicators for use in a larger composite indicator project. The processed data will later contribute towards calculating a comparative SDG readiness score across countries.

The following datasets are obtained through: https://unstats.un.org/sdgs/dataportal/ 

**Disclaimer: This dataset has already been standardised by the UN SDG database, therefore some common data cleaning steps such as formatting corrections, unit conversions, and major transformations were not required. The cleaning process mainly focused on handling missing values, identifying duplicates, selecting relevant indicators and preparing the data for analysis.**

### 1.1 Import Data

```python
# Import pandas
import pandas as pd

# Load datasets
infant_df = pd.read_excel("SH_DYN_IMRT.xlsx")
disease_df = pd.read_excel("SH_DTH_NCOM.xlsx")
suicide_df = pd.read_excel("SH_STA_SCIDE.xlsx")

# Display first 5 rows of each dataframe
print("Infant Mortality Dataset:")
display(infant_df.head())

print("\nNon-Communicable Diseases Dataset:")
display(disease_df.head())

print("\nSuicide Rates Dataset:")
display(suicide_df.head())
```

#### Meaning of Datasets

- infant_df : Infant mortality rate (deaths per 1,000 live births) SH_DYN_IMRT
- disease_df : Mortality rate attributed to cardiovascular disease, cancer, diabetes or chronic respiratory disease (probability) SH_DTH_NCOM
- suicide_df : Suicide mortality rate, by sex (deaths per 100,000 population) SH_STA_SCIDE

### 2.1 Select Columns

```python
# Select columns for each dataset

# Years
years_2000_2024 = [str(year) for year in range(2000, 2025)]

# Clean column names
infant_df.columns = infant_df.columns.str.strip()
disease_df.columns = disease_df.columns.str.strip()
suicide_df.columns = suicide_df.columns.str.strip()

# Select only years that exist in each dataset
infant_years = [year for year in years_2000_2024 if year in infant_df.columns]
disease_years = [year for year in years_2000_2024 if year in disease_df.columns]
suicide_years = [year for year in years_2000_2024 if year in suicide_df.columns]

infant_df = infant_df[['GeoAreaName', 'Sex'] + infant_years]
disease_df = disease_df[['GeoAreaName', 'Sex'] + disease_years]
suicide_df = suicide_df[['GeoAreaName', 'Sex'] + suicide_years]

# Display first 5 rows
print("Infant Dataset:")
display(infant_df.head())

print("\nDisease Dataset:")
display(disease_df.head())

print("\nSuicide Dataset:")
display(suicide_df.head())
```

#### 2.2 Selection of Columns

At this stage, only the variables required for the **SDG 3 sub-index analysis** are retained from each dataset.

The selection keeps:

* all countries available within the datasets,
* identifying columns such as `GeoAreaName` and `Sex`,
* yearly health indicator values from 2000 up to the most recent year available within each dataset.

For the SDG 3 sub-index, the selected datasets include infant mortality, disease-related indicators, and suicide mortality indicators.
