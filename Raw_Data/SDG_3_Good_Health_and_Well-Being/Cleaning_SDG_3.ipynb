# Sustainable Development Goal 3: Good Health and Well-Being

This Jupyter Notebook focuses on the development and preparation of the SDG 3 Sub-Index, which relates to the United Nations Sustainable Development Goal 3: Good Health and Well-Being.

The purpose of this notebook is to import, inspect, clean and prepare datasets associated with health and well-being indicators for use in a larger composite indicator project. The processed data will later contribute towards calculating a comparative SDG readiness score across countries.

The following datasets are obtained through: https://unstats.un.org/sdgs/dataportal/ 

**Disclaimer: This dataset has already been standardised by the UN SDG database, therefore some common data cleaning steps such as formatting corrections, unit conversions and major transformations were not required. The cleaning process mainly focused on handling missing values, identifying duplicates, selecting relevant indicators and preparing the data for analysis.**

### 1.1 Import Data

```python
# Import pandas
import pandas as pd

# Load datasets
infant_df = pd.read_excel("SH_DYN_IMRT.xlsx")
disease_df = pd.read_excel("SH_DTH_NCOM.xlsx")
suicide_df = pd.read_excel("SH_STA_SCIDE.xlsx")

# Display first 5 rows of each dataframe
print("Infant Mortality Dataset:")
display(infant_df.head())

print("\nNon-Communicable Diseases Dataset:")
display(disease_df.head())

print("\nSuicide Rates Dataset:")
display(suicide_df.head())
```

#### Meaning of Datasets

- infant_df : Infant mortality rate (deaths per 1,000 live births) SH_DYN_IMRT
- disease_df : Mortality rate attributed to cardiovascular disease, cancer, diabetes or chronic respiratory disease (probability) SH_DTH_NCOM
- suicide_df : Suicide mortality rate, by sex (deaths per 100,000 population) SH_STA_SCIDE

### 2.1 Select Columns

```python
# Select columns for each dataset

# Years
years_2000_2024 = [str(year) for year in range(2000, 2025)]

# Clean column names
infant_df.columns = infant_df.columns.str.strip()
disease_df.columns = disease_df.columns.str.strip()
suicide_df.columns = suicide_df.columns.str.strip()

# Select only years that exist in each dataset
infant_years = [year for year in years_2000_2024 if year in infant_df.columns]
disease_years = [year for year in years_2000_2024 if year in disease_df.columns]
suicide_years = [year for year in years_2000_2024 if year in suicide_df.columns]

infant_df = infant_df[['GeoAreaName', 'Sex'] + infant_years]
disease_df = disease_df[['GeoAreaName', 'Sex'] + disease_years]
suicide_df = suicide_df[['GeoAreaName', 'Sex'] + suicide_years]

# Display first 5 rows
print("Infant Dataset:")
display(infant_df.head())

print("\nDisease Dataset:")
display(disease_df.head())

print("\nSuicide Dataset:")
display(suicide_df.head())
```

#### 2.2 Selection of Columns

At this stage, only the variables required for the **SDG 3 sub-index analysis** are retained from each dataset.

The selection keeps:

* all countries available within the datasets,
* identifying columns such as `GeoAreaName` and `Sex`,
* yearly health indicator values from 2000 up to the most recent year available within each dataset.

For the SDG 3 sub-index, the selected datasets include infant mortality, disease-related indicators and suicide mortality indicators.

### 3.1 Dealing with Duplicates

Duplicates will be dealt in 4.1 Check & Handling of Missing Values (Imputation of Missing Data)

#### infant_df, disease_df, suicide_df

During the inspection of the SDG 3 datasets, duplicate country entries were identified across all three datasets. These duplicates were not considered data errors, as they occur because the indicators are separated by biological sex categories.

Each country contains three records:

* `MALE`
* `FEMALE`
* `BOTHSEX`

The `BOTHSEX` category represents the combined or averaged value across both male and female populations. As a result, countries appear multiple times within the datasets, with each row representing a different sex classification rather than a true duplicate observation.

This structure is consistent across the infant mortality, disease and suicide datasets.

### 5.1 Check for Outliers

### infant_df

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,6))

infant_df[years_2000_2024].boxplot()

plt.xticks(rotation=90)
plt.ylabel("Infant Mortality Rate")
plt.title("Boxplot of Infant Mortality Rate by Year")

plt.show()
```

```python
infant_possible_outliers = []

for year in years_2000_2024:

    # Skip missing year columns
    if year not in infant_df.columns:
        continue

    # Calculate 95th percentile threshold
    upper_cap = infant_df[year].quantile(0.95)

    # Get values above 95th percentile
    outliers = infant_df[
        infant_df[year] > upper_cap
    ][['GeoAreaName', 'Sex', year]].copy()

    outliers = outliers.rename(
        columns={year: 'InfantMortalityRate'}
    )

    outliers['Year'] = year
    outliers['95thPercentileCap'] = upper_cap

    infant_possible_outliers.append(outliers)

infant_possible_outliers_df = pd.concat(infant_possible_outliers)

print("Values Above 95th Percentile Before Capping:")
display(infant_possible_outliers_df)

# -----------------------------
# Apply 95th percentile capping
# -----------------------------
infant_df_capped = infant_df.copy()

for year in years_2000_2024:

    # Skip missing year columns
    if year not in infant_df_capped.columns:
        continue

    # Calculate 95th percentile
    upper_cap = infant_df_capped[year].quantile(0.95)

    # Cap values above threshold
    infant_df_capped[year] = infant_df_capped[year].clip(
        upper=upper_cap
    )

infant_df = infant_df_capped

# Boxplot after capping
plt.figure(figsize=(15, 6))

infant_df.boxplot(
    column=years_2000_2024,
    rot=90
)

plt.title('Boxplot of Capped (95th Percentile) Infant Mortality Rate by Year')
plt.xlabel('Year')
plt.ylabel('Infant Mortality Rate')

plt.grid(True)

plt.show()
```

The infant mortality dataset contained some very high values, mainly from countries in Sub-Saharan Africa and countries affected by conflict or instability, such as Afghanistan. These values are not mistakes, as they reflect real health challenges caused by factors such as poor healthcare access, poverty, conflict, malnutrition and disease.

The boxplots also show that these extreme values appear consistently across many years, rather than being random one-off spikes. Most countries are grouped within a much lower range, while a small number of countries sit far above the upper whiskers as clear outliers. In my opinion, this suggests that the dataset is heavily right-skewed and that a few countries could disproportionately influence the analysis if left untreated.

For this reason, I capped values above the 95th percentile for each year. This keeps the countries in the dataset, but reduces the effect of extreme values so that the index is not dominated by a small number of countries.


### disease_df

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,6))

disease_df[disease_years].boxplot()

plt.xticks(rotation=90)
plt.ylabel("Mortality Probability")
plt.title("Boxplot of Mortality Rate from Non-Communicable Diseases by Year")

plt.show()
```

```python
disease_possible_outliers = []

for year in disease_years:

    Q1 = disease_df[year].quantile(0.25)
    Q3 = disease_df[year].quantile(0.75)

    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = disease_df[
        (disease_df[year] < lower_bound) |
        (disease_df[year] > upper_bound)
    ][['GeoAreaName', year]].copy()

    outliers = outliers.rename(columns={year: 'MortalityProbability'})

    outliers['Year'] = year

    disease_possible_outliers.append(outliers)

disease_possible_outliers_df = pd.concat(disease_possible_outliers)

display(disease_possible_outliers_df)
```

The outliers within the non-communicable disease mortality dataset were reviewed before deciding whether treatment was necessary. Most of the higher values came from countries such as Afghanistan, Kazakhstan, Russian Federation and Ukraine. In my opinion, these values are realistic and reflect genuine health issues within those countries rather than errors in the dataset.

Unlike the infant mortality dataset, the distribution here is much more balanced and the outliers are not extremely far from the rest of the data. Because of this, I decided not to cap or remove them. Keeping these values helps preserve meaningful differences between countries, especially for countries with higher risks linked to smoking, alcohol consumption, cardiovascular disease, conflict, or weaker healthcare systems.

Since the overall distribution still remains relatively stable, I felt that treating the outliers would reduce the realism of the dataset more than it would improve it.


### suicide_df

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,6))

suicide_df[suicide_years].boxplot()

plt.xticks(rotation=90)
plt.ylabel("Suicide Mortality Rate (per 100,000 population)")
plt.title("Boxplot of Suicide Mortality Rate by Year")

plt.show()
```

```python
suicide_possible_outliers = []

for year in suicide_years:

    Q1 = suicide_df[year].quantile(0.25)
    Q3 = suicide_df[year].quantile(0.75)

    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = suicide_df[
        (suicide_df[year] < lower_bound) |
        (suicide_df[year] > upper_bound)
    ][['GeoAreaName', year]].copy()

    outliers = outliers.rename(columns={year: 'MortalityRate'})

    outliers['Year'] = year

    suicide_possible_outliers.append(outliers)

suicide_possible_outliers_df = pd.concat(suicide_possible_outliers)

display(suicide_possible_outliers_df)
```

These outliers look much more reasonable after filtering and cleaning the dataset. The countries appearing most often, such as Lithuania, Russian Federation, Belarus, Kazakhstan and Sri Lanka, are countries that have historically recorded high suicide mortality rates.

In my opinion, these values should be kept and not capped. Unlike the infant mortality dataset, the suicide mortality distribution still looks relatively controlled and the outliers are not extremely far away from the rest of the data. The higher values reflect genuine social and public health issues such as mental health challenges, alcohol abuse, economic instability and limited access to mental health support.

I think capping these values would reduce meaningful differences between countries. Since the distribution remains fairly stable overall, I believe keeping the outliers provides a more accurate representation of global suicide mortality patterns.


### 6.1 Creating New Variables

#### infant_df

```python
import pandas as pd
import numpy as np

# Select 5-year intervals + latest year
infant_forecast_years = ['2005', '2010', '2015', '2020', '2024']

# Temporary forecasting dataframe
infant_forecast_df = infant_df[
    ['GeoAreaName', 'Sex'] + infant_forecast_years
].copy()

# Keep only total sex value to avoid Male/Female duplicates
infant_forecast_df = infant_forecast_df[
    infant_forecast_df['Sex'] == 'BOTHSEX'
]

# Years for regression
x_years = np.array([2005, 2010, 2015, 2020, 2024])

# Store predictions
predictions_2030 = []

for index, row in infant_forecast_df.iterrows():

    y_values = row[infant_forecast_years].values.astype(float)

    # Linear trend
    slope, intercept = np.polyfit(x_years, y_values, 1)

    # Predict 2030
    predicted_2030 = slope * 2030 + intercept

    predictions_2030.append(predicted_2030)

# Create final dataset
infant_2030_df = pd.DataFrame({
    'GeoAreaName': infant_forecast_df['GeoAreaName'],
    'InfantMortality_2030_Predicted': predictions_2030
})

# Display result
display(infant_2030_df)
```

#### disease_df

```python
# Select 5-year intervals + latest year
disease_forecast_years = ['2005', '2010', '2015', '2020', '2021']

# Temporary forecasting dataframe
disease_forecast_df = disease_df[
    ['GeoAreaName', 'Sex'] + disease_forecast_years
].copy()

# Keep only total sex values
disease_forecast_df = disease_forecast_df[
    disease_forecast_df['Sex'] == 'BOTHSEX'
]

# Years for regression
x_years = np.array([2005, 2010, 2015, 2020, 2021])

# Store predictions
predictions_2030 = []

for index, row in disease_forecast_df.iterrows():

    y_values = row[disease_forecast_years].values.astype(float)

    # Linear trend
    slope, intercept = np.polyfit(x_years, y_values, 1)

    # Predict 2030
    predicted_2030 = slope * 2030 + intercept

    predictions_2030.append(predicted_2030)

# Create final dataset
disease_2030_df = pd.DataFrame({
    'GeoAreaName': disease_forecast_df['GeoAreaName'],
    'DiseaseMortality_2030_Predicted': predictions_2030
})

# Display result
display(disease_2030_df)
```

#### suicide_df

```python
# Temporary forecasting dataframe
suicide_forecast_df = suicide_df[
    ['GeoAreaName', 'Sex'] + disease_forecast_years
].copy()

# Keep only total sex values
suicide_forecast_df = suicide_forecast_df[
    suicide_forecast_df['Sex'] == 'BOTHSEX'
]

# Store predictions
predictions_2030 = []

for index, row in suicide_forecast_df.iterrows():

    y_values = row[disease_forecast_years].values.astype(float)

    # Linear trend
    slope, intercept = np.polyfit(x_years, y_values, 1)

    # Predict 2030
    predicted_2030 = slope * 2030 + intercept

    predictions_2030.append(predicted_2030)

# Create final dataset
suicide_2030_df = pd.DataFrame({
    'GeoAreaName': suicide_forecast_df['GeoAreaName'],
    'SuicideMortality_2030_Predicted': predictions_2030
})

# Display result
display(suicide_2030_df)
```

To estimate future SDG 3 readiness, I followed a similar forecasting approach to the one used for SDG 6. I created projected 2030 variables for each SDG 3 indicator using historical trend data.

The indicators included:

- Infant Mortality Rate
- Disease Mortality Rate
- Suicide Mortality Rate

For this section, I selected 5-year interval observations:

- 2005
- 2010
- 2015
- 2020

The main reason for using 5-year intervals was that the disease and suicide datasets were only available in wider time intervals rather than complete yearly data. Because of this, I decided to apply the same interval structure across the SDG 3 indicators for consistency.

In my opinion, this was the fairest approach because it keeps all SDG 3 indicators aligned under the same forecasting method. Even though the infant mortality dataset had more yearly data available, using the same 5-year structure prevents one indicator from being treated differently from the others.

I also included the most recent available year for each dataset:

- 2024 for infant mortality
- 2021 for disease mortality
- 2021 for suicide mortality

This helped make the 2030 projections more realistic by including the latest available observed values before forecasting.

A linear regression trend was then calculated individually for each country, and this trend was used to estimate the projected indicator value for 2030.

Countries with missing values across the selected forecasting years were excluded from prediction to avoid creating unreliable estimates from incomplete data.

The final forecasting datasets only retained:

- `GeoAreaName`
- the projected 2030 indicator variable

Retained in the following:

- infant_2030_df
- disease_2030_df
- suicide_2030_df

This simplified structure was used because these projected variables will later be merged together when constructing the SDG 3 sub-index.

### 7.1 Multivariate Analysis

#### 7.2 Combining Data

```python
# Merge datasets and keep all countries from all datasets
# Missing indicator values are filled with 0 to represent no available readiness evidence for that indicator

sdg3_combined_df = (
    infant_2030_df
    .merge(disease_2030_df, on='GeoAreaName', how='outer')
    .merge(suicide_2030_df, on='GeoAreaName', how='outer')
)

# Replace missing values with 0
sdg3_combined_df = sdg3_combined_df.fillna(0)

display(sdg3_combined_df)
print(f"Final dataset shape: {sdg3_combined_df.shape}")
```

When combining the SDG 3 datasets, I decided to use an outer join instead of an inner join because not every country had data available across all three health indicators. Some countries may have data for infant mortality but not for disease mortality or suicide mortality. This is common in international datasets where reporting standards, healthcare systems, and data collection capabilities vary between countries.

If I had used an inner join, only countries that existed in all three datasets would remain in the final combined dataset. In my opinion, this would remove too many countries and discard potentially useful information. A country may still provide meaningful evidence of health readiness through one or two indicators, even if another indicator is missing.

For this reason, I chose to use an outer join so that all countries appearing in at least one SDG 3 dataset would still be included in the final sub-index dataset. This creates a more inclusive dataset and avoids excluding countries entirely due to incomplete reporting.

After merging the datasets, I filled all missing values with `0`. I made this decision because a missing value can reasonably represent the absence of measurable readiness for that particular health indicator within the context of the sub-index. Rather than removing the country completely, assigning a value of `0` allows the country to still participate in the overall SDG 3 readiness assessment while reflecting that no supporting evidence exists for the missing indicator.

This approach also helps preserve balance within the sub-index calculation. Countries with strong performance in some health indicators can still achieve a reasonable SDG 3 readiness score, even if one dataset is unavailable. At the same time, the missing indicator still lowers the overall score appropriately because a value of `0` contributes negatively to readiness.

#### 7.3 Heatmap

```python
import matplotlib.pyplot as plt
import seaborn as sns

sdg3_numeric = sdg3_combined_df.select_dtypes(include='number')

correlation_matrix = sdg3_numeric.corr()

# Plot heatmap
plt.figure(figsize=(10, 8))

sns.heatmap(
    correlation_matrix,
    annot=True,
    cmap='coolwarm',
    fmt=".2f",
    linewidths=0.5
)

plt.title("SDG 3 Correlation Heatmap")
plt.xticks(rotation=45)
plt.yticks(rotation=0)

plt.show()
```

The correlation heatmap for the SDG 3 indicators shows mostly low to moderate relationships between the variables, with no extremely high correlation values present.

- Infant Mortality vs Disease Mortality: **0.58**
- Infant Mortality vs Suicide Mortality: **0.05**
- Disease Mortality vs Suicide Mortality: **0.33**

In my opinion, these results are still acceptable for a sub-index because the indicators are not strongly correlated to the point where serious multicollinearity becomes a concern.

The moderate correlation of **0.58** between infant mortality and disease mortality is understandable because both indicators are associated with overall national health conditions and healthcare quality. Countries with weaker healthcare systems may naturally experience higher values across both indicators. However, the relationship is still below the level typically associated with problematic multicollinearity.

The remaining correlations are relatively weak, particularly the relationship between infant mortality and suicide mortality (**0.05**), which suggests that suicide mortality captures a different dimension of SDG 3 performance compared to the other health indicators.

Overall, the heatmap suggests that the SDG 3 indicators remain sufficiently independent from one another, meaning the sub-index is less likely to be distorted by redundant variables or excessive overlap between indicators.
